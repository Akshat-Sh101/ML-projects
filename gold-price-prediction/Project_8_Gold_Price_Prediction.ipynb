**Work Flow**

Gold Data -> Data Preprocessing -> Data Analysis -> Train_test_split -> Random Forest Regression model -> Evaluaiton

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

**Data Collection and Presprocessing**

```python
dataset = pd.read_csv('/content/gld_price_data.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# getting some statistical measures of the data
dataset.describe()
```

**Correlation**

*   Positive Correlation
*   Negative Correlation



```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
correlation = dataset.corr()
```

```python
# Visualizing the correlation in a heatmap
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
# Correlation of all columns WRT gold(GLD)
print(correlation['GLD'])
```

```python
# checking the distribution of gold prices
sns.displot(dataset['GLD'],color='green')
```

**Spliting the data and label**

```python
x = dataset.drop(columns=['GLD','Date'],axis =1)
y = dataset['GLD']
```

```python
print(x)
print(y)
```

**Train_Test_Split**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

**Model Training**

### Random Forest Regression Model

```python
model = RandomForestRegressor(n_estimators=100)
```

```python
model.fit(x_train,y_train)
```

**Model Evaluation**

```python
x_test_predict = model.predict(x_test)
x_test_score = metrics.r2_score(y_test,x_test_predict)
print("R squared error : ",x_test_score)
```

**Comparing the Actual Values with Predicted Values in a plot**

```python
y_test = list(y_test)
```

```python
plt.plot(y_test,color='blue',label='Actual Value')
plt.plot(x_test_predict,color='green',label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
